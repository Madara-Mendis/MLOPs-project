```python
def get_product(a: int, b: int) -> int:
    return a * b    
```

```python
get_product(3,4)
```

```python
get_product(3,"4")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(a: int, b: int) -> int:
    return a * b
```

```python
get_product(3,4)
```

```python
get_product(3,"4")
```
